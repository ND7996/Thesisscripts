```python
from Bio import SeqIO
import re

fh = open('GPX_human.fa','w')
for seq_record in SeqIO.parse("glutathione.faa", "fasta"):
    print(seq_record.id)
    item = seq_record.description
    HScode = 'OX=9606'
    match = re.search(HScode, item)
    if match:
        print('Yes!')
        print(item.split())
        print(len(seq_record))
        fh.write(seq_record.format("fasta"))

fh.close()
```
